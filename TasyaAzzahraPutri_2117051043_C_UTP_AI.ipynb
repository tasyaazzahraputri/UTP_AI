## **Library**

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

## **Load Data**

```python
# Load data dari Github

dataset = pd.read_csv("https://raw.githubusercontent.com/tasyaazzahraputri/UTP_AI/main/Iris.csv")
dataset
```

## **Preprocessing**

```python
# data cleaning
dataset = dataset.dropna()
dataset = dataset.drop(["Id"], axis = 1)
dataset
```

```python
# type conversion
dataset.loc [dataset['Species'] == 'Iris-setosa', 'Species'] = '1'
dataset.loc [dataset['Species'] == 'Iris-versicolor', 'Species'] = '2'
dataset.loc [dataset['Species'] == 'Iris-virginica', 'Species'] = '3'
dataset
```

## **Splitting Data**

```python
data_train, data_test = train_test_split(dataset, train_size = 0.8, random_state = 65)
```

```python
X_train = data_train.drop('Species', axis=1)
y_train = data_train['Species']
X_test = data_test.drop('Species', axis=1)
y_test = data_test['Species']
```

```python
# Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## **Model**

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

## **Evaluasi**

```python
akurasi = model.score(X_test, y_test)
akurasi
```

## **Prediksi Data Dummy**

```python
dummy = pd.read_csv("https://raw.githubusercontent.com/tasyaazzahraputri/UTP_AI/main/data_dummy.csv")
dummy
```

```python
prediksi_spesies = model.predict(dummy.values)
dummy["Spesies"] = prediksi_spesies
dummy
```
